## 長期の均衡解を求める

```python
E_list = [5, 8, 10]
Col_list = [5, 8, 10, 20, 30]
for Col in Col_list:
    for E in E_list:
        print("="*40)
        print(f"E={E}, K={Col}^2")
        
        K = Col * Col
    
        M = 0.1 * K
        N = 0.1 * K

        # Scaling * tが格子の最小距離となる。
        Scaling = 300 / Col
        alter_T_num = 0.5

        city_network = lattice.make_lattice(Col)

        # 通勤費用パラメータ
        t = 0.1

        # 距離抵抗パラメータ
        tau = 0.01

        # 総土地供給量
        S_total = K
        S_bar = S_total / K

        Coordinate_Data = \
        np.array([(city_network['node_dic'][str(i)]['x_pos']*Scaling,
                   city_network['node_dic'][str(i)]['y_pos']*Scaling) for i in range(K)])
        distance_matrix = distance.squareform(distance.pdist(Coordinate_Data))

        theta_firm = 1.0
        theta_house = 1.0

        alpha_1 = 0.4
        alpha_2 = 0.4
        beta_1 = 0.4
        beta_2 = 0.4

        L = 0.2
        eta = 1.2
        p_proj = 1e-3

        RW_ini = 1.0
        RW_proj = 1e-3
        
        m_per = M / K
        m0 = np.array([m_per] * K)
        n0 = np.full((K, K), N / (K * K))
        
        method = "FISTA"
        dic = "Scaling=300divCol"
        
        #パラメータ設定
        prm = exp_Short_FISTA.Parameter(
                    Col, distance_matrix, t, tau, Scaling, S_total, S_bar,
                    theta_firm, theta_house, E, RW_proj, alter_T_num,
                    alpha_1, alpha_2, beta_1, beta_2, M, N)
        
        algprm = exp_Short_FISTA.Algo_Parameter(L, eta, p_proj)
        short = exp_Short_FISTA.Short(prm, algprm)
        long = exp_Long_FISTA.Long(prm, algprm, short)
        
        start_time = time.process_time()
        m, n, RW, long_iteration, obj_list = long.solve(m0, n0, RW_ini, long_itr=100000, err_long=1e-3)
        end_time = time.process_time()
        elapsed_time = end_time - start_time
        
        print("経過時間:", elapsed_time)
        
        print("len(iteration):", len(long_iteration))
        print("len(obj_list):", len(obj_list))
        exp_kakunou.long_kakunou_val(long_iteration, obj_list, prm, algprm, method, dic)
```
