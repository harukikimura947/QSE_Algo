```python
tau_list = [0.01]
Col_list = [15]
theta_list = [0.5]
for Col in Col_list:
    for tau in tau_list:
        for theta in theta_list:
            print("="*40)
            print(f"theta={theta}, Scaling={Scaling}, tau={tau}")

            Col = 20
            K = Col * Col

            M = 0.1 * K
            N = 0.1 * K

            # Scaling * tが格子の最小距離となる。
            Scaling = Scaling
            alter_T_num = 0.5

            city_network = lattice.make_lattice(Col)

            # 通勤費用パラメータ
            t = 0.1

            # 距離抵抗パラメータ
            tau = tau
            
            E = 5

            # 総土地供給量
            S_total = K
            S_bar = S_total / K

            Coordinate_Data = \
            np.array([(city_network['node_dic'][str(i)]['x_pos']*Scaling,
                       city_network['node_dic'][str(i)]['y_pos']*Scaling) for i in range(K)])
            distance_matrix = distance.squareform(distance.pdist(Coordinate_Data))

            theta_firm = theta
            theta_house = theta

            alpha_1 = 0.4
            alpha_2 = 0.4
            beta_1 = 0.4
            beta_2 = 0.4

            L = 0.2
            eta = 1.2
            p_proj = 1e-3

            RW_ini = 1.0
            RW_proj = 1e-3

            err_short = 1e-5
            err_long = "200iteration"

            m_per = M / K
            m0 = np.array([m_per] * K)
            n0 = np.full((K, K), N / (K * K))

            method = "FISTA"
            dic = "Scaling=15"

            #パラメータ設定
            prm = exp_parameter.Parameter(
                        Col, distance_matrix, t, tau, Scaling, S_total, S_bar,
                        theta_firm, theta_house, E, RW_proj, alter_T_num,
                        alpha_1, alpha_2, beta_1, beta_2, M, N)

            algprm = exp_parameter.Algo_Parameter(L, eta, p_proj)
            short = exp_Short_FISTA.Short(prm, algprm)
            long = exp_Long_FISTA.Long(prm, algprm, short)
            
#             dF = pd.read_excel(fr'C:/Users/kimura/OneDrive - 国立大学法人東北大学/numerical/short_dual/exp/excelfile/' \
#                    fr'shortlong/{method}/err_short={err_short}/err_long={err_long}/{dic}/RW_proj={prm.RW_proj}/alter_T_num={prm.alter_T_num}/' \
#                    fr'L={algprm.L}/eta={algprm.eta}/p_proj={algprm.p_proj}/E={prm.E}/theta={prm.theta_firm}/K={prm.Col}^2/truevalue/jikken_F.xlsx')
            
#             dH = pd.read_excel(fr'C:/Users/kimura/OneDrive - 国立大学法人東北大学/numerical/short_dual/exp/excelfile/' \
#                    fr'shortlong/{method}/err_short={err_short}/err_long={err_long}/{dic}/RW_proj={prm.RW_proj}/alter_T_num={prm.alter_T_num}/' \
#                    fr'L={algprm.L}/eta={algprm.eta}/p_proj={algprm.p_proj}/E={prm.E}/theta={prm.theta_firm}/K={prm.Col}^2/truevalue/jikken_H.xlsx')
        
#             m_corr = np.array(dF['m'].tolist())
#             n_corr = np.array(dH['n'].tolist())
#             R_corr = np.array(dF['R'].tolist())
#             W_corr = np.array(dF['W'].tolist())

#             mn_corr = long.bond(m_corr, n_corr)
#             RW_corr = np.concatenate((R_corr, W_corr))

#             obj_corr = long.Z_LP(mn_corr, RW_corr)

            obj_corr = 1.0
            
            start_time = time.process_time()
            m, n, RW, long_iteration, obj_list, obj_rel_list = long.solve(m0, n0, RW_ini, err_short, err_long, obj_corr, long_itr=200)
            end_time = time.process_time()
            elapsed_time = end_time - start_time

            print("CPU_time:", elapsed_time, "秒")

            R = RW[:prm.K]
            W = RW[prm.K:]

            pi = short.pi(R, W, m)
            v = short.v(R, W)
            
            S_H_hat, L_H_hat, S_F, L_F = short.demand_hat(R, W)

            # err_long = "obj_rel_smaller_1e-4"

            exp_kakunou.long_kakunou_true(m, n, R, W, pi, v, prm, algprm, method, err_short, err_long, dic)
            exp_kakunou.long_kakunou_nume_num(long_iteration, prm, algprm, method, err_short, err_long, dic)
            # exp_kakunou.long_kakunou_rel(obj_rel_list, prm, algprm, method, err_short, err_long, dic)
            exp_kakunou.long_kakunou_val(long_iteration, obj_list, prm, algprm, method, err_short, err_long, dic)
            
            vmin = 0
            vmax = prm.S_bar

            mat = np.reshape(m * S_F, (Col, Col))

            plt.figure(figsize=(5,5))
            plt.imshow(mat, interpolation='nearest', vmin = vmin, vmax = vmax, cmap='seismic')

            plt.colorbar()

            plt.xlabel('m * S_F', fontsize=20)

            plt.gca().invert_yaxis()
            plt.xticks([])
            plt.yticks([])

            plt.show()
            
            matrix = np.reshape(m * S_F, (Col, Col))

            # 行列の座標を作成
            x, y = np.meshgrid(range(Col), range(Col))

            # 3Dグラフの作成
            fig = plt.figure()
            ax = fig.add_subplot(111, projection='3d')

            ax.plot_wireframe(x, y, matrix, cmap='viridis')

            # グラフの軸ラベル設定
            ax.set_xlabel('Col')
            ax.set_ylabel('Row')
            ax.set_zlabel('m * S_F')

            plt.gca().invert_xaxis()

            # グラフ表示
            plt.show()
```
