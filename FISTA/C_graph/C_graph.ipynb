```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV
file_path = "C://Users//kimura//Downloads//short_solve//data//K=2500//par_L=0.0050,eta=1.5000//iteration_data.csv"

# CSV
df = pd.read_csv(file_path)

# 
plt.figure(figsize=(12, 8))

# max_diff 
plt.subplot(2, 2, 1)
plt.semilogy(df['iteration'], df['max_diff'], label='max_diff')
plt.title('max_diff')
plt.xlabel('Iteration')
plt.ylabel('max_relative_diff')
plt.grid(True)

# R_max_diff 
plt.subplot(2, 2, 2)
plt.semilogy(df['iteration'], df['R_max_diff'], label='R_max_diff', color='orange')
plt.title('R_max_diff')
plt.xlabel('Iteration')
plt.ylabel('max_relative_diff')
plt.grid(True)

# W_max_diff 
plt.subplot(2, 2, 3)
plt.semilogy(df['iteration'], df['W_max_diff'], label='W_max_diff', color='green')
plt.title('W_max_diff')
plt.xlabel('Iteration')
plt.ylabel('max_relative_diff')
plt.grid(True)

# Z_diff 
plt.subplot(2, 2, 4)
plt.semilogy(df['iteration'], df['Z_diff'], label='Z_diff', color='red')
plt.title('Z_diff')
plt.xlabel('Iteration')
plt.ylabel('relative_diff')
plt.grid(True)

# 
plt.tight_layout()

# 
plt.show()
```
