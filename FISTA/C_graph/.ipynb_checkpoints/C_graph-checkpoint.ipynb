```python
# plotter.py (コマンドライン引数を使わないバージョン)
import matplotlib.pyplot as plt
import pandas as pd
import os

# パラメータ設定 (C言語側の設定と合わせる)
K = 900  # 例
par_L = [0.0050, 0.0080, 0.0100, 0.0200, 0.0300, 0.0500, 0.0800]
eta = [1.5000, 1.8000, 2.0000]

# Windows のユーザーディレクトリを正確に取得
username = os.environ.get("USERNAME") or os.environ.get("USER")  # Windows/Linux対応
if username is None:
    raise RuntimeError("Error: USERNAME could not be determined.")

base_dir = os.path.join("C:\\Users", username, "Downloads", "short_sovle", "data")

for i in par_L:
    for j in eta:

        # CSVファイルのフルパスを構築
        csv_filepath = os.path.join(base_dir, f"K={K}", f"par_L={i:.4f},eta={j:.4f}", "iteration_data.csv")

        # CSVファイルを読み込む
        if not os.path.exists(csv_filepath):
            print(f"Error: {csv_filepath} not found.")
            print("Make sure the CSV file exists at the specified path.")
            continue  # プログラム終了せずに次のループへ

        try:
            df = pd.read_csv(csv_filepath)
        except pd.errors.EmptyDataError:
            print(f"Error: {csv_filepath} is empty.")
            continue
        except Exception as e:
            print(f"An unexpected error occurred: {e}")
            continue

        # グラフを描画
        plt.plot(df["iteration"], df["max_diff"])
        plt.plot(df["iteration"], df["ave_diff"])
        plt.plot(df["iteration"], df["max_diff_true"])
        plt.plot(df["iteration"], df["ave_dff_true"])
        plt.xlabel("Iteration")
        plt.ylabel("Max Difference")
        plt.title("par_L={i:.4f},eta={j:.4f}")
        plt.grid(True)
        plt.yscale("log")
        plt.show()
```
